```python
def avg_2_mod(x, y):
    if (x == 20):
        return np.NaN # np.nan or np.NAN
    else:
        return (x + y) / 2
```

```python
avg_2_mod(20, 100)
```

```python
avg_2_mod(5, 10)
```

```python
avg_2_mod(df['a'], df['b'])
```

```python
import numpy as np
```

```python
avg_2_mod_vec = np.vectorize(avg_2_mod)
```

```python
avg_2_mod_vec(df['a'], df['b'])
```

```python
@np.vectorize
def avg_2_mod(x, y):
    if (x == 20):
        return np.NaN # np.nan or np.NAN
    else:
        return (x + y) / 2
```

```python
avg_2_mod(df['a'], df['b'])
```

```python
import numba
```

```python
@numba.vectorize
def avg_2_mod_numbda(x, y):
    if (x == 20):
        return np.NaN # np.nan or np.NAN
    else:
        return (x + y) / 2
```

```python
avg_2_mod_numbda(df['a'].values, df['b'].values)
```

```python
%%timeit
avg_2_mod(df['a'], df['b'])
```

```python
%%timeit
avg_2_mod_numbda(df['a'].values, df['b'].values)
```
